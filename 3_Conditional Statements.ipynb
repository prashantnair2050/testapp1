```python
# 1. if statement -- Deal with one condition

temp = 9

if temp < 10:
  print("Low Temp")
  print("Test")
```

```python
# 2. # 2. if..else statement --- Binary Condition
temp=10


if temp < 10:
  print("Low Temp")
else:
  print("High Temp")
```

```python
# 3. elif ladder --- Used to deal with more than 2 conditions

temp=10


if temp < 10:
  print("Low Temp")

elif temp >=10 and temp < 15:
  print("Moderate Temp")
else:
  print("High Temp")
```

```python

if temp < 10:
  print("Low Temp")

elif temp >=10 and temp < 15:
  print("Moderate Temp")
elif temp >=15:
  print("High Temp")
```

```python
# Ternary Operator ---- Short hand operator of if else

num1 = 10

if num1 % 2 == 0:
  print("Even")
else:
  print("Odd")
```

```python
"Even" if num1 % 2 == 0 else "Odd"
```
